```
metric = []
accuracy = []
n_trial = 5
filter_factor = np.linspace(2, 5, 16)

for i in filter_factor:
    metric_value = 0
    accuracy_value = 0

    for _ in range(n_trial): 
        filter = (abs(data.pmra - center_x) < i*std) & (abs(data.pmdec - center_y) < i*std)

        working_data = data[filter]
        features = working_data.loc[:,['pmra', 'pmdec']]

        scaled_features = pd.DataFrame({})
        for column in features.columns:
            scaled_features[column] = (features[column] - np.median(features[column]))/np.std(features[column])

        from sklearn.mixture import GaussianMixture
        gmm = GaussianMixture(n_components=2)
        gmm.fit(scaled_features)

        #predictions from gmm
        labels = gmm.predict(scaled_features)

        probs = gmm.predict_proba(scaled_features)

        working_data.loc[:, 'PMemb'] = probs[:, 0]
        working_data.loc[:, 'gmm_label'] = labels

        non_member = working_data[working_data['PMemb'] <= 0.2]
        member = working_data[working_data['PMemb'] >= 0.8]

        if np.std(member.pmra) > np.std(non_member.pmra):
            member = working_data[working_data['PMemb'] <= 0.2]
            non_member = working_data[working_data['PMemb'] >= 0.8]

        metric_1 = (np.std(non_member.pmra) - np.std(member.pmra)) / np.std(member.pmra)
        metric_2 = (np.std(non_member.pmdec) - np.std(member.pmdec)) / np.std(member.pmdec)

        metric_value += np.mean([metric_1, metric_2])
        
        true_positive = sum(member.label == 1)/(len(member)+1)
        true_negative = sum(non_member.label == 0)/(len(non_member)+1)
        accuracy_value += np.mean([true_positive, true_negative])

    
    metric.append(metric_value/n_trial)
    accuracy.append(accuracy_value/n_trial)

    plt.figure(figsize=(22,6))

    plt.suptitle('Filter - Center: '+ str(i))

    plt.subplot(131)
    sns.scatterplot(data = working_data, x='pmra', y='pmdec', hue = 'gmm_label')

    plt.subplot(132)
    sns.scatterplot(data = member, x='pmra', y='pmdec', label = 'member')
    sns.scatterplot(data = non_member, x='pmra', y='pmdec', label = 'non_member')

    plt.subplot(133)
    sns.scatterplot(data = working_data, x='pmra', y='pmdec', hue = 'label')
    plt.show()
```

```

plt.figure(figsize=(14,6))

plt.subplot(121)
sns.lineplot(x=filter_factor, y=metric, label = 'Metric')

plt.subplot(122)
sns.lineplot(x=filter_factor, y=accuracy, label = 'Accuracy')
plt.show()
```

### Radial Density Plot

```
def get_distance_from_center(cluster_data, center):
    data = np.array(cluster_data.loc[:, ['pmra', 'pmdec']])

    dist_array = data - list(center)

    dist = ((dist_array[:, 0]*np.cos(np.radians(dist_array[:, 1])))**2 + 
                   dist_array[:, 1]**2)**0.5

    return dist
```

```
dist = get_distance_from_center(data, center)
counts, bins, bars = plt.hist(dist, bins=[0.5*np.sqrt(i) for i in range(100)])
```

```
sns.histplot(dist, bins=[0.5*np.sqrt(i) for i in range(100)], color='tab:green')
plt.xlabel('Euclidean Distance in pm space')
plt.title('Star count around the pm center for regions with equal area')
plt.plot(bins[1:], counts, color = 'orange')
plt.show()
```

```
plt.plot(bins[1:], counts)
plt.xlabel('Euclidean Distance in PM space')
plt.ylabel('Star Count')
plt.title('Star count around the pm center for regions with equal area')
plt.show()
```
